```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

### Hello Clustering

```python
"""
Steps:
1: Initialization -> Pick K random points as K centroid locations
2: Assign -> For each data point, find the closet centroid and assign that data point to the centroid.
3: Update -> For each centroid, calculate the maean from the data points assigned to it.
4: Repeat -> repeat step 2 and 3 until cluster centroid is stable.
"""
```

```python
# JUST Plot function
def plot(data, set_cluster, K_cord, colors):
    plt.figure(figsize=(7,7))
    labels = [i+1 for i in range(len(colors))]
    ax = plt.subplot(1, 1, 1)
    data_x = data[:, 0]
    data_y = data[:, 1]
        
    for i in range(len(data)):
        ax.scatter(data_x[i], data_y[i], marker="o", color=colors[set_cluster[i]], s=100, edgecolors='black')
    
    for k in range(K):
        kx = K_cord[k][0]
        ky = K_cord[k][1]
        ax.scatter(kx, ky, marker="*", color=colors[k], s=400, edgecolors='black', label = labels[k])
    ax.legend()
    plt.show()
```

```python
# Assign data point to cluster
def euclid_dist(X, point_k):
    dist = np.sqrt(np.sum(np.square(X-point_k), axis=1))
    return dist

def Assign():
    all_k_dist = []
    for k in range(len(K_cord)):
        centroid = K_cord[k]
        distance = euclid_dist(data, centroid)
        all_k_dist.append(distance)

    set_cluster = np.argmin(all_k_dist, axis=0)
    plot(data, set_cluster, K_cord, colors)
    return set_cluster
```

```python
def Update():
    # Update cluster centroid
    for k in range(K):
        where = np.squeeze(np.argwhere(set_cluster==k))
        new_cord = np.mean(data[where], axis=0)
        K_cord[k] = new_cord
    plot(data, set_cluster, K_cord, colors)
```

```python
# Cost function
def cost_function(data, K_cord):
    data_x = data[:, 0]
    data_y = data[:, 1]
    cost = 0
    for k in range(len(K_cord)):
        X = np.squeeze(data[np.argwhere(set_cluster==k)])
        dist = np.sqrt(np.sum(np.square(X-K_cord[k])))
        cost += dist
    return cost
```

# T1. SET K = 3

```python
######################## SETUP ###############################
# x and y coordinate.
data = np.array([[1,2],[3,3],[2,2],[8,8],[6,6],[7,7],[-3,-3],[-7,-7],[-2,-4]])
data_x = data[:, 0]
data_y = data[:, 1]
set_cluster = np.zeros(len(data), dtype=int)

# T1: starting point K=3 --> (3,3) (2,2) (-3,-3)

# Set starting point for cluser centroid
K = 3
K_cord = np.array([[3,3],[2,2],[-3,-3]], dtype=float)
colors = ['red', 'yellow', 'blue']
plot(data, set_cluster, K_cord, colors)
```

```python
######################## ITERATE ###############################
for i in range(3):
    print("Iteration: ", i+1)
    print("Assign Phase . . .")
    set_cluster = Assign()
    print("Update centroid Phase . . .")
    Update()
    print("Cost Function = ", cost_function(data, K_cord), "\n")
```
