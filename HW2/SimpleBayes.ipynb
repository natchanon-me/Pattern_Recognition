```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import math
%matplotlib inline
```

```python
mu1 = 3
variance1 = 2
mu2 = -3
variance2 = 2
sigma1 = math.sqrt(variance1)
sigma2 = math.sqrt(variance2)
x1 = np.linspace(mu1 - 3*sigma1, mu1 + 3*sigma1, 1000)
x2 = np.linspace(mu2 - 3*sigma2, mu2 + 3*sigma2, 1000)
```

```python
def solve(m1,m2,std1,std2):
    a = 1/(2*std1**2) - 1/(2*std2**2)
    b = m2/(std2**2) - m1/(std1**2)
    c = m1**2 /(2*std1**2) - m2**2 / (2*std2**2) - np.log(std2/std1)
    return (np.roots([a,b,c]))

result = solve(mu1, mu2, sigma1, sigma2)
result= np.around(result[0], decimals=3)

ax = plt.subplot(1, 1, 1)
ax.plot(x1, norm.pdf(x1, mu1, sigma1))
ax.plot(x2, norm.pdf(x2, mu2, sigma2))
ax.plot(result, norm.pdf(result, mu1, sigma1),'o')
ax.axvline(x=result, color='r', linestyle='--')
ax.annotate(str(result), xy=(result, norm.pdf(result, mu1, sigma1)))
plt.show()
```

```python
# If prior is assigned
prior1 = 0.8
prior2 = (1-prior1)

result = 1.93
ax2 = plt.subplot(1, 1, 1)
ax2.plot(x1, norm.pdf(x1, mu1, sigma1)*prior1)
ax2.plot(x2, norm.pdf(x2, mu2, sigma2)*prior2)
ax2.plot(result, norm.pdf(result, mu2, sigma2)*prior2,'o')
ax2.axvline(x=result, color='r', linestyle='--')
ax2.annotate(str(result), xy=(result, norm.pdf(result, mu1, sigma1)))
plt.show()
```

```python
mu1 = 5
variance1 = 2
mu2 = 0
variance2 = 4
sigma1 = math.sqrt(variance1)
sigma2 = math.sqrt(variance2)
x1 = np.linspace(mu1 - 3*sigma1, mu1 + 3*sigma1, 1000)
x2 = np.linspace(mu2 - 3*sigma2, mu2 + 3*sigma2, 1000)

result = solve(mu1, mu2, sigma1, sigma2)
result= np.around(result[1], decimals=3)

ax4 = plt.subplot(1, 1, 1)
ax4.plot(x1, norm.pdf(x1, mu1, sigma1))
ax4.plot(x2, norm.pdf(x2, mu2, sigma2))
ax4.plot(result, norm.pdf(result, mu1, sigma1),'o')
ax4.axvline(x=result, color='r', linestyle='--')
ax4.annotate(str(result), xy=(result, norm.pdf(result, mu1, sigma1)))
plt.show()
```
